
Importando e Usando o Fake

Criando Dados de Clientes

```python
#Importar biblioteca 
from faker import Faker

# Criando uma instância do Faker
fake = Faker("pt_BR")  

# Gerando um conjunto de dados fictícios
dados_clientes = []

for _ in range(100):  # Gera 100 registros
    dados_clientes.append({
        'Nome': fake.name(),
        'Endereço': fake.address(),
        'Email': fake.email(),
            })

# Exibindo os dados gerados
for cliente in dados_clientes:
    print(cliente)
```

Simulação de Transações:

```python
from faker import Faker
import random

fake = Faker("pt_BR")
products = ['Produto A', 'Produto B', 'Produto C']

for _ in range(100):
    print(fake.date_time_this_year(), random.choice(products), round(random.uniform(10.0, 100.0), 2))
```

Dicas para Dados Realistas

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os dados com distribuição normal
values = np.random.normal(loc=50, scale=10, size=100)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.hist(values, bins=15, color='skyblue', edgecolor='black')
plt.title('Distribuição Normal dos Valores')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

 # "Refinando Dados Fictícios com Python 

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from faker import Faker
import random

fake = Faker()
data = [{"name": fake.name(), "price": round(random.uniform(10, 100), 2)} for _ in range(100)]

def generate_dates(start_date, n):
    return [start_date + timedelta(days=i) for i in range(n)]

dates = generate_dates(datetime(2023, 1, 1), 30)
plt.plot(dates, range(30), marker='o')
plt.title('Datas de Compras')
plt.xlabel('Data')
plt.ylabel('Número de Compras')
plt.xticks(rotation=45)
plt.show()
```

 Relacionamentos entre Variáveis e Gráficos de Dispersão

```python
quantidades = [random.randint(1, 10) for _ in range(100)]
preços = [data[i]['price'] for i in range(100)]

plt.scatter(preços, quantidades)
plt.title('Relação Preço vs. Quantidade Vendida')
plt.xlabel('Preço')
plt.ylabel('Quantidade Vendida')
plt.show()
```

Distribuição Realista de Valores com Seaborn

```python
import seaborn as sns

preços_normal = [round(random.gauss(50, 15), 2) for _ in range(100)]
sns.histplot(preços_normal, bins=15, kde=True)
plt.title('Distribuição de Preços')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

 Inclusão de Anomalias com Boxplots

```python
prices_with_anomalies = preços_normal+ [150, 200, -20]  # Adicionando anomalias
sns.boxplot(x=prices_with_anomalies)
plt.title('Boxplot com Anomalias')
plt.xlabel('Preço')
plt.show()
```

# Criando um Dataset e Realizando Análises



Criando um Dataset Fictício

```python
from faker import Faker
import pandas as pd
import random

fake = Faker()

# Configurações
num_records = 1000
categories = ['Eletrônicos', 'Roupas', 'Alimentos', 'Móveis']

# Gerando dados
data = {
    'ID': [i for i in range(1, num_records + 1)],
    'Produto': [fake.word() for _ in range(num_records)],
    'Categoria': [random.choice(categories) for _ in range(num_records)],
    'Preço': [round(random.uniform(10, 500), 2) for _ in range(num_records)],
    'Data da Venda': [fake.date_this_year() for _ in range(num_records)],
    'Cliente': [fake.name() for _ in range(num_records)],
}

df = pd.DataFrame(data)
df.to_csv('vendas_ficticias.csv', index=False)
print('Dataset gerado com sucesso!')
```

```python
import matplotlib.pyplot as plt

vendas_por_categoria = df['Categoria'].value_counts()

plt.figure(figsize=(10, 6))
vendas_por_categoria.plot(kind='bar', color='skyblue')
plt.title('Número de Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Número de Vendas')
plt.show()
```

```python
df['Data da Venda'] = pd.to_datetime(df['Data da Venda'])
vendas_por_data = df.groupby(df['Data da Venda'].dt.date).size()

plt.figure(figsize=(12, 6))
vendas_por_data.plot(kind='line', marker='o', color='orange')
plt.title('Tendência de Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Número de Vendas')
plt.grid()
plt.show()
```
