```python
#Importar biblioteca 
from faker import Faker

# Criando uma instância do Faker
fake = Faker("pt_BR")  

# Gerando um conjunto de dados fictícios
dados_clientes = []

for _ in range(100):  # Gera 100 registros
    dados_clientes.append({
        'Nome': fake.name(),
        'Endereço': fake.address(),
        'Email': fake.email(),
            })

# Exibindo os dados gerados
for cliente in dados_clientes:
    print(cliente)
```

 # "Refinando Dados Fictícios com Python 

```python
from faker import Faker
import random

fake = Faker()
data = [{"name": fake.name(), "price": round(random.uniform(10, 100), 2)} for _ in range(100)]
```

Refinamento de Padrões Temporais com Matplotlib

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

def generate_dates(start_date, n):
    return [start_date + timedelta(days=i) for i in range(n)]

dates = generate_dates(datetime(2023, 1, 1), 30)
plt.plot(dates, range(30), marker='o')
plt.title('Datas de Compras')
plt.xlabel('Data')
plt.ylabel('Número de Compras')
plt.xticks(rotation=45)
plt.show()
```

 Relacionamentos entre Variáveis e Gráficos de Dispersão

```python
quantidades = [random.randint(1, 10) for _ in range(100)]
preços = [data[i]['price'] for i in range(100)]

plt.scatter(preços, quantidades)
plt.title('Relação Preço vs. Quantidade Vendida')
plt.xlabel('Preço')
plt.ylabel('Quantidade Vendida')
plt.show()
```

Distribuição Realista de Valores com Seaborn

```python
import seaborn as sns

preços_normal = [round(random.gauss(50, 15), 2) for _ in range(100)]
sns.histplot(preços_normal, bins=15, kde=True)
plt.title('Distribuição de Preços')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

 Inclusão de Anomalias com Boxplots

```python
prices_with_anomalies = preços_normal+ [150, 200, -20]  # Adicionando anomalias
sns.boxplot(x=prices_with_anomalies)
plt.title('Boxplot com Anomalias')
plt.xlabel('Preço')
plt.show()
```

# Criado com Dados Fictícios para Próximos Projeto"

```python
# Importando as bibliotecas

from faker import Faker
import random
from datetime import datetime, timedelta
fake = Faker()

# Geração de Clientes e Produtos 

customers = [fake.name() for _ in range(100)]  # 100 clientes fictícios
products = ['Camiseta', 'Calça', 'Tênis', 'Jaqueta', 'Bolsa']
prices = [round(random.uniform(50, 500), 2) for _ in range(len(products))]  # Preços dos produtos

# Refinando a Data da Venda e Quantidades 
def generate_business_days(start_date, n):
    return [start_date + timedelta(days=random.randint(1, 5)) for _ in range(n)]

purchase_dates = generate_business_days(datetime(2023, 1, 1), 200)  # 200 vendas fictícias
quantities = [random.randint(1, 5) for _ in range(200)]  # Quantidade por venda

# Relacionando Preço, Quantidade e Valor Total

sales_data = []
for i in range(200):
    product = random.choice(products)
    price = prices[products.index(product)]
    quantity = quantities[i]
    total_value = round(price * quantity, 2)
    sales_data.append({"Cliente": random.choice(customers), 
                       "Produto": product, 
                       "Preço": price, 
                       "Quantidade": quantity, 
                       "Valor Total": total_value, 
                       "Data da Compra": purchase_dates[i]})
    

# Visualizando os Dados com Gráficos

import matplotlib.pyplot as plt
import seaborn as sns

prices_list = [item['Preço'] for item in sales_data]
quantities_list = [item['Quantidade'] for item in sales_data]
total_values_list = [item['Valor Total'] for item in sales_data]

sns.histplot(prices_list, bins=10, kde=True)
plt.title('Distribuição de Preços dos Produtos')
plt.show()

sns.histplot(total_values_list, bins=10, kde=True)
plt.title('Distribuição de Valores Totais por Venda')
plt.show()

```

```python
# Inserção de Anomalias
sales_data.append({"Cliente": fake.name(), "Produto": "Smartphone", "Preço": 5000, 
                   "Quantidade": 10, "Valor Total": 50000, "Data da Compra": datetime(2023, 6, 1)})

sales_data.append({"Cliente": fake.name(), "Produto": "Camiseta", "Preço": -50, 
                   "Quantidade": 1, "Valor Total": -50, "Data da Compra": datetime(2023, 7, 1)})


print(sales_data)
```

```python
#salvando o dataset sales_date

import pandas as pd

# Cria o DataFrame a partir de sales_data
df_sales = pd.DataFrame(sales_data)

# Salva o DataFrame em um arquivo CSV
df_sales.to_csv('sales_data.csv', index=False)

print("Arquivo 'sales_data.csv' salvo com sucesso!")

```
